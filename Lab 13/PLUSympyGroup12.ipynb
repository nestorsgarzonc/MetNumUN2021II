```python
import sympy 
import numpy as np
def PLU(A,k):
  
  N = A.shape[0]
  u = A.copy()
  L = sympy.eye(N)
  P = sympy.eye(N) 
  
  for j in range(k):
      lamb = sympy.eye(N)
      perm = sympy.eye(N)
      row = j
      max = u[row, j]
      for i in range(j, N, 1):
        if abs(u[i, j]) > abs(max): 
          max = u[i, j]
          row = i    
      swp_r = perm[row,:].copy()
      perm[row,:] = perm[j,:]
      perm[j,:] =  swp_r
      u = perm @ u 
      gamma = u[j+1:, j] / u[j, j]
      lam[j+1:, j] = -gamma
      u = lam @ u
      P = perm @ P
      L = L @ perm @ lam 
  return P,P@L,u
```
